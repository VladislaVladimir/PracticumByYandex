## Предобработка данных

Проверим фрейм на наличие явных дубликатов

```python
data.duplicated().sum()
```

Отлично, дубликатов нет!

### Обрботка пропусков

   Высота потолков, общее количество этажей в доме, площадь квартиры, площадь кухни указаны не везде. Скорее всего человек, подававший объявление, не посчитал нужным заполнить данный раздел. Так как дальше эти столбцы будут использоваться в расчетах, то оставим пропуски без изменения, так как заполнение пропусков усредненными значениями заведомо даст ложные граффики и диаграммы. Будем рабоатать с теми значениями, которые есть. Однако, заменим пропуски на unknown в столбце locality_name, для того, чтобы наглядно показать неизвестность названия населенного пункта.

```python
data['locality_name'] = data['locality_name'].fillna('unknown')
```

Обработаем пропуски в столбце `balcony`. Если человек не указал балконы, то скорее их всего нет. В связи с этим заменим пропуски на значение 0.

```python
data['balcony'] = data['balcony'].fillna(0)
```

В столбце `is_apartment`, так же присутствуют пропуски. Заполненые строки имеют значение True, то есть это булев массив. Однако, так как имеются пропуски NaN, тип данной строки определется как object. Для того, чтобы превартить ее в bool заменим пропуски на False, т.е. "не аппартаменты".

```python
data['is_apartment']=data['is_apartment'].fillna(False)
```

### Смена типа данных

Изменим формат данных в столбце `rooms` с float на int так как количество комнат  не может быть дробным числом.

```python
data['rooms'] = data['rooms'].astype('int')
```

### Смена формата даты

Уберем значение времени в столбце `first_day_exposition`, так как оно не несет никакой смысловй нагрузки для дальнейшей работы. Будем работать исключительно с числом, месяцем и годом.

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%d')
```

### Вывод

В данном разделе провели предобработку данных: 

1) Нашли и заполнили пропуски

2) Сменили формат данных на в столбце rooms с float на int, по причине того, что количесвто комнат не может быть дробным числом.

3) Так как в исследовании мы будем работать с числом, месяцем и годом, то показатель времени решено убрать.


## Расчёты и добавление результатов в таблицу

### Средняя стоимость квадратного метра

Рассчитаем среднюю стоимость квадратного метра и добавим новый столбец в таблицу:

```python
data['price_for_meter'] = data['last_price']/data['total_area']
```

### Группировка по этажам

Разобьем по категориям этаж квартиры. Если квартира не первом этаже присвоем категорию "первый", если на последнем -"последний", в остальных случаях выводится категория "другой".
Для присвоения категорий напишем функцию:

```python
def group_floor(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
```

Добавим новый столбец применив функцию к таблице.

```python
data['group_floor'] = data.apply(group_floor, axis = 1)
```

### Разбивка по дням недели, месяцам и годам

Разобьем даты публикации отдельно на дни недели, недели, месяца и года и добавим эти значеняи в таблицу:

```python
data['day'] = data['first_day_exposition'].dt.weekday
```

```python
data['week'] = data['first_day_exposition'].dt.week
```

```python
data['month'] = data['first_day_exposition'].dt.month
```

```python
data['year'] = data['first_day_exposition'].dt.year
```

### Соотношение жилой и общей площади, а также отношение площади кухни к общей.

```python
data['dep_living_area']=data['living_area']/data['total_area'] #отноешение жилой площади к общей площади
```

```python
data['dep_kitchen_area'] = data['kitchen_area']/data['total_area'] #отношение площади  кухни к общей площади
```

### Вывод

В данном разделе произведены следущие операции:
1) Провели расчеты средней цены квадратного метра на недвижимости и добавли значения в наш датасет

2) Провели группировку этажей по категориям "последний", "первый" и "другой"

3) Произвели разбиву даты по дням недели, месяцам и годам и дабавлили получившиеся значения в датасет

4) Рсчитали и добавили в датасет отношение жилой площади и площади кухни к общей площади.

### Обрезаем хвосты

Обрежем выбросы:

```python
# Выкинем из расчетов недвижимость обладющую более 8 комнатами
query_table = data.query('rooms<=8')
# Масимальную общую площадь недвижимотси ограничим 200 кв.м. минимальная 12 (что скорее всего комната в камуналке)
query_table = query_table.query('total_area<=200')
# Примим срок продажи квартиры от 7 до 500 дней.
query_table = query_table.query('days_exposition>=7 & days_exposition<=800')
# Обрежем значения меньше миллиона рублей и значения превышающие 40 миллионов рублей
query_table = query_table.query('last_price>=1000000 & last_price<=40000000')
# Выкинем знаячения высоты потолков меньше 2 метров и больше 4 метров
query_table = query_table.query('ceiling_height>=2 & ceiling_height<4')
```

### 10 населённых пунктов с наибольшим числом объявлений 

Для начала составим список из 10 самых популярных населенных пунктов встречающихся в объявлениях и сохраним в переменную ten_popular_lacation. 

```python
ten_popular_lacation = (query_table['locality_name']
                        .value_counts()
                        .sort_values(ascending=False).head(10).index
                       )
ten_popular_lacation
```

Далее создадим срез в которых присутсвуют значения из ten_popular_lacation в query_table, затем сгруппируем все в сводную таблицу, отсортируем и выведем на экран.


```python
ten_popular_lacation_by_meter = (query_table
                                 .query('locality_name in @ten_popular_lacation')
                                 .pivot_table(index='locality_name', values='price_for_meter', aggfunc='mean')
                                 .sort_values(by = 'price_for_meter', ascending = False)
                                 .reset_index()
)  #помним про бритву Оккама для красивого кода:)
```
